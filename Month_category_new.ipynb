```python
import pandas as pd
```

```python
t1 = pd.read_csv('Content.csv')
t2 = pd.read_csv('Reactions.csv')
t3 = pd.read_csv('ReactionTypes.csv')
```

```python
t1['Category'] = t1['Category'].str.replace('"', '')
t1['Category'] = t1['Category'].str.lower().str.strip()
t1 = t1.drop(columns=['User ID', 'URL', 'Unnamed: 0'])
t1 = t1.rename(columns={'Type': 'Content Type'})
t1.head()
```

```python
#t2 = t2.drop(columns=['Unnamed: 0'])
t2 = t2.dropna(subset=['Type'])
t2['User ID'] = t2['User ID'].fillna('Unknown')
t2 = t2.rename(columns={'Type': 'Reaction Type'})
t2['Datetime'] = pd.to_datetime(t2['Datetime']) 
t2['Month'] = t2['Datetime'].dt.month
t2['Hour'] = t2['Datetime'].dt.hour

def categorize_time_of_day(hour):
    if 3 <= hour <= 5:
        return ' Early Morning'
    elif 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 19:
        return 'Evening'
    elif 20 <= hour <= 23:
        return 'Night'
    else:
        return 'Late Night'

t2['Time of Day'] = t2['Hour'].apply(categorize_time_of_day)
t2.info()
```

```python
t3 = t3.drop(columns=['Unnamed: 0'])
t3 = t3.rename(columns={'Type': 'Reaction Type'})
t3 = t3.rename(columns={'Score': 'Reaction Popularity'})
t3.head()
```

```python
t2_t1 = pd.merge(t2, t1, on='Content ID', how='left')
all_three = pd.merge(t2_t1, t3, on='Reaction Type', how='left')
all_three.head()
final_result = all_three[['Category', 'Reaction Popularity', 'Month']]
final_result.to_csv('monthwise_output_jupyter.csv')
final_result.info()
```

```python
t1.to_csv('t1.csv', index = False)
t2.to_csv('t2.csv', index = False)
t3.to_csv('t3.csv', index = False)
```
